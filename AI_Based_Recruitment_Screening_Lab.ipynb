## Step 3: Understand the Dataset Shape and Columns

```python
# This cell checks the size of the dataset and the column names.
# shape tells us how many rows and columns are present.
# columns tells us the names of all variables in the dataset.

print("Dataset shape:", df.shape)
print("Column names:", list(df.columns))
```

## Step 4: Basic Data Inspection

```python
# This cell displays the first few rows of the dataset.
# head() is useful for quickly checking whether the data looks correct.

df.head()
```

## Step 5: Define a Job Description

```python
# This cell defines the target job description.
# In a real recruitment system, this text would come from an actual job posting.
# We keep it short and relevant for learning purposes.

job_description = "python sql machine learning nlp communication dashboards recruitment analytics"

print("Job Description:")
print(job_description)
```

## Step 6: Convert Resume Skills and Job Description into Numerical Form Using TF-IDF

```python
# This cell uses TF-IDF Vectorizer.
# TF-IDF stands for Term Frequency - Inverse Document Frequency.
# It converts text into numbers based on the importance of words.
# We combine the job description and candidate skills in one list
# so that both can be represented in the same vector space.

documents = [job_description] + df["skills"].tolist()

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

## Step 7: Calculate Cosine Similarity Between Job Description and Candidate Skills

```python
# This cell calculates cosine similarity.
# Cosine similarity measures how similar two text vectors are.
# A value closer to 1 means higher similarity.
# A value closer to 0 means lower similarity.

job_vector = tfidf_matrix[0:1]
candidate_vectors = tfidf_matrix[1:]

similarity_scores = cosine_similarity(job_vector, candidate_vectors).flatten()

df["similarity_score"] = similarity_scores

print("Cosine similarity scores calculated successfully.")
df[["candidate_id", "name", "skills", "similarity_score"]]
```

## Step 8: Normalize Experience and Certifications for Fair Scoring

```python
# This cell scales experience and certifications to a 0 to 1 range.
# Normalization is important because experience and certifications
# are measured on different numeric scales.

df["experience_score"] = df["experience_years"] / df["experience_years"].max()
df["certification_score"] = df["certifications"] / df["certifications"].max()

df[["name", "experience_years", "experience_score", "certifications", "certification_score"]]
```

## Step 9: Convert Education into a Simple Numerical Score

```python
# This cell maps education levels into a simple score.
# This is only for demonstration.
# In real hiring, education should be interpreted carefully and ethically.

education_map = {
    "B.Com": 0.50,
    "BBA": 0.55,
    "BCA": 0.55,
    "B.Sc": 0.60,
    "B.Tech": 0.70,
    "MBA": 0.75,
    "MBA HR": 0.80,
    "MCA": 0.80,
    "M.Sc": 0.80,
    "M.Tech": 0.90,
    "M.Stat": 0.90
}

df["education_score"] = df["education"].map(education_map)

df[["name", "education", "education_score"]]
```

## Step 10: Build a Weighted Candidate Score

```python
# This cell creates a final weighted score.
# We assign:
# 50% weight to similarity score
# 25% weight to experience score
# 15% weight to education score
# 10% weight to certification score
#
# These weights are only for teaching purposes.
# Organizations may define different weights based on role requirements.

df["final_score"] = (
    0.50 * df["similarity_score"] +
    0.25 * df["experience_score"] +
    0.15 * df["education_score"] +
    0.10 * df["certification_score"]
)

df[["name", "similarity_score", "experience_score", "education_score", "certification_score", "final_score"]]
```

## Step 11: Rank Candidates from Highest to Lowest Score

```python
# This cell sorts the candidates by final score in descending order.
# Highest score appears first, which means better fit according to the model.

ranked_df = df.sort_values(by="final_score", ascending=False).reset_index(drop=True)

ranked_df[["candidate_id", "name", "skills", "experience_years", "education", "certifications", "final_score"]]
```

## Step 12: Create a Shortlist of Top 5 Candidates

```python
# This cell selects the top 5 candidates.
# These are the candidates most aligned with the job description
# according to the simple model created in this lab.

top_5 = ranked_df.head(5)

top_5[["candidate_id", "name", "skills", "experience_years", "education", "certifications", "final_score"]]
```

## Step 13: Add HR Interpretation

```python
# This cell adds a recommendation label.
# It converts numeric scores into simple decision categories.

def label_candidate(score):
    if score >= 0.65:
        return "Strong Match"
    elif score >= 0.50:
        return "Moderate Match"
    else:
        return "Low Match"

ranked_df["recommendation"] = ranked_df["final_score"].apply(label_candidate)

ranked_df[["candidate_id", "name", "final_score", "recommendation"]]
```

## Step 14: View Final Recruitment Output Table

```python
# This cell shows the final output table.
# This output simulates the screening result that HR could review.

final_output = ranked_df[[
    "candidate_id",
    "name",
    "skills",
    "experience_years",
    "education",
    "certifications",
    "similarity_score",
    "experience_score",
    "education_score",
    "certification_score",
    "final_score",
    "recommendation"
]]

final_output
```
